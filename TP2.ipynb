```python
import numpy as np
from tabulate import tabulate

# TP2 optimisation descente de gradient

## f(x) = x2 + sin(x)
## Descente de gradient à pas fixe alpha = 0.01
### fonction coût
def g(x):
    return x**2 + np.sin(x)

### gradient de la fonction
def grad_f(x):
    return 2*x + np.cos(x)

def gradient_descent_fixed_step(x_initial, alpha, stop_criterion):
    x = x_initial
    results = []
    for i in range(stop_criterion):
        ## Calcul du gradient de la fonction coût
        grad = grad_f(x)
        x = x - alpha * grad
        # Stockage de l'état à chaque itération
        results.append([i+1, x, g(x)])
    return x, results

## Définition des paramètres
initialisation = [10, 20, 30]
alpha = 0.01
stop_criterion = 30

## Exécution de la descente et affichage des résultats
table_data = []
for x_initial in initialisation:
    optimal_x, results = gradient_descent_fixed_step(x_initial, alpha, stop_criterion)
    table_data.extend(results)
    print(f"Optimal x for initial x = {x_initial}: {optimal_x}")

print("\nRésultats sous forme de tableau :")
print(tabulate(table_data, headers=["Iteration", "x", "f(x)"]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# TP2 optimisation descente de gradient

## Descente de gradient à pas fixe alpha = 0.01
### Fonction coût
def g(x):
    return x**2 + np.sin(x)

### Gradient de la fonction
def grad_g(x):
    return 2*x + np.cos(x)

def gradient_descent_fixed_step(x_initial, alpha, stop_criterion):
    x = x_initial
    x_values = [x]
    for i in range(stop_criterion):
        ## Calcul du gradient de la fonction coût
        grad = grad_g(x)
        x = x - alpha * grad
        x_values.append(x)
    return x, x_values

## Définition des paramètres
initialisations = [10, 20, 30]
alpha = 0.01
stop_criterion = 30

## Exécution de la descente et collecte des résultats
results = {}
for x_initial in initialisations:
    optimal_x, x_values = gradient_descent_fixed_step(x_initial, alpha, stop_criterion)
    results[x_initial] = x_values

## Création du graphique
x_range = np.linspace(-35, 35, 400)
y_range = g(x_range)

plt.figure(figsize=(12, 8))
plt.plot(x_range, y_range, label='g(x) = x^2 + sin(x)', color='blue')

for x_initial, x_values in results.items():
    plt.scatter(x_values, g(np.array(x_values)), label=f'Initial x = {x_initial}', zorder=5)
    plt.plot(x_values, g(np.array(x_values)), linestyle='--')

plt.title('Descente de Gradient à Pas Fixe pour Différents Points de Départ')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fonction h(x)
def h(x):
    return x**4 / 4 - x**3 / 3 - x**2 + 1

# Gradient de la fonction h(x)
def grad_h(x):
    return x**3 - x**2 - 2*x

# Descente de gradient à pas fixe
def gradient_descent_fixed_step(x_initial, alpha, stop_criterion):
    x = x_initial
    x_values = [x]
    for i in range(stop_criterion):
        # Calcul du gradient de la fonction coût
        grad = grad_h(x)
        x = x - alpha * grad
        x_values.append(x)
    return x, x_values

# Définition des paramètres
initialisations = [1, 2, 3]  # Réduire les valeurs initiales pour éviter les grands nombres
alpha = 0.01
stop_criterion = 30

# Exécution de la descente et collecte des résultats
results = {}
for x_initial in initialisations:
    optimal_x, x_values = gradient_descent_fixed_step(x_initial, alpha, stop_criterion)
    results[x_initial] = x_values

# Création du graphique
x_range = np.linspace(-5, 5, 400)  # Limiter la plage d'affichage
y_range = h(x_range)

plt.figure(figsize=(12, 8))
plt.plot(x_range, y_range, label='h(x) = x^4/4 - x^3/3 - x^2 + 1', color='blue')

for x_initial, x_values in results.items():
    plt.scatter(x_values, h(np.array(x_values)), label=f'Initial x = {x_initial}', zorder=5)
    plt.plot(x_values, h(np.array(x_values)), linestyle='--')

plt.title('Descente de Gradient à Pas Fixe pour Différents Points de Départ')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# TP2 optimisation descente de gradient

## Fonction coût
def f(x, y):
    return (x - 2)**2 + (y - 3)**2

## Gradient de la fonction
def grad_f(x, y):
    grad_x = 2 * (x - 2)
    grad_y = 2 * (y - 3)
    return grad_x, grad_y

def gradient_descent_fixed_step_2d(x_initial, y_initial, alpha, stop_criterion):
    x, y = x_initial, y_initial
    points = [(x, y)]
    for i in range(stop_criterion):
        ## Calcul du gradient de la fonction coût
        grad_x, grad_y = grad_f(x, y)
        x = x - alpha * grad_x
        y = y - alpha * grad_y
        points.append((x, y))
    return x, y, points

## Définition des paramètres
initialisations = [(10, 10), (20, 20), (30, 30)]
alpha = 0.1
stop_criterion = 30

## Exécution de la descente et collecte des résultats
results = {}
for (x_initial, y_initial) in initialisations:
    optimal_x, optimal_y, points = gradient_descent_fixed_step_2d(x_initial, y_initial, alpha, stop_criterion)
    results[(x_initial, y_initial)] = points

## Création du graphique
x_range = np.linspace(-10, 40, 400)
y_range = np.linspace(-10, 40, 400)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)

colors = ['r', 'g', 'b']
for idx, ((x_initial, y_initial), points) in enumerate(results.items()):
    points = np.array(points)
    ax.scatter(points[:, 0], points[:, 1], f(points[:, 0], points[:, 1]), color=colors[idx], label=f'Initial (x, y) = ({x_initial}, {y_initial})', zorder=5)
    ax.plot(points[:, 0], points[:, 1], f(points[:, 0], points[:, 1]), linestyle='--', color=colors[idx])

ax.set_title('Descente de Gradient à Pas Fixe pour Différents Points de Départ')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.legend()
plt.show()
```
